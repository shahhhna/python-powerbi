```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna().copy()
```

```python
binary_cols = [
    "Partner","Dependents","PhoneService",
    "PaperlessBilling","Churn"
]

for col in binary_cols:
    df[col] = df[col].map({"No":0, "Yes":1})
```

```python
df
```

## Feature Engineering

**Tenure Groups (Lifecycle Segmentation)**

```python

def tenure_group(x):
    if x <= 6:
        return "0-6 Months"
    elif x <= 24:
        return "6-24 Months"
    else:
        return "24+ Months"

df['TenureGroup'] = df['tenure'].apply(tenure_group)
```

**Revenue Buckets (Value Segmentation)**

```python
def revenue_bucket(x):
    if x <= 50:
        return "Low"
    elif x <= 80:
        return "Medium"
    else:
        return "High"

df['RevenueBucket'] = df['MonthlyCharges'].apply(revenue_bucket)
df
```

**Contract Flag**

```python
df['IsMonthly'] = df['Contract'].apply(
    lambda x: 1 if x == "Month-to-month" else 0
)
```

## Section 3 – Revenue Impact

**Calculate Total MRR**

```python
mrr = df['MonthlyCharges'].sum()
mrr
#total monthly recurring revenue.
```

**Revenue Lost Due to Churn**

```python
revenue_lost = df[df['Churn'] == 1]['MonthlyCharges'].sum()
revenue_lost
#how much recurring revenue you lose every month.
```

**Annual Revenue Loss**

```python
annual_revenue_lost = revenue_lost * 12
annual_revenue_lost
#The company is losing approximately $X annually due to churn.
```

**Revenue Lost by Segment**

```python
#Revenue loss by Contract

df[df['Churn'] == 1].groupby('Contract')['MonthlyCharges'].sum()

#Revenue loss by TenureGroup

df[df['Churn'] == 1].groupby('TenureGroup')['MonthlyCharges'].sum()

#Which segment causes the biggest financial damage
```

**5% Churn Reduction Scenario**

```python
#If churn reduces by 5%, calculate:

customers_saved = len(df) * 0.05
avg_revenue = df['MonthlyCharges'].mean()

monthly_gain = customers_saved * avg_revenue
annual_gain = monthly_gain * 12

#A 5% churn reduction could generate approximately $X additional annual recurring revenue.
```

```python
import re

def clean_column(name):
    # Add underscore between lower and upper case letters
    name = re.sub(r'([a-z0-9])([A-Z])', r'\1_\2', name)
    
    # Convert to title case (first letter capital for each word)
    name = name.title()
    
    return name

df.columns = [clean_column(col) for col in df.columns]
```
